## Filtering DataFrames

```python
titanic = pd.read_csv("data/titanic.csv")
```

```python
mask1 = titanic.sex == "male"
```

```python
mask2 = titanic.age > 14
```

```python
mask1.head()
```

```python
(mask1 & mask2).head()
```

```python
male_survivors = titanic.loc[mask1 & mask2, ["survived", "pclass", "sex", "age"]]
```

```python
male_survivors.head(40)
```

```python
male_survivors.info()
```

```python
male_survivors.describe()
```

```python
titanic.describe()
```

## OR Conditions

```python
mask3 = titanic.sex == "female"
```

```python
mask4 = titanic.age < 14
```

```python
(mask3 | mask4).head()
```

```python
titanic.loc[mask3|mask4]
```

```python
women_or_children = titanic.loc[mask3 | mask4, ["survived", "pclass", "sex", "age"]]
```

```python
women_or_children.head()
```

```python
women_or_children.info()
```

```python
women_or_children.describe()
```

## Advanced Filtering with between and isin and ~

```python
summer = pd.read_csv("data/summer.csv")
```

```python
og_1988 = summer.loc[summer.Year == 1988]
```

```python
og_1988.head()
```

```python
og_1988.info()
```

```python
og_since1992 = summer.loc[summer.Year >= 1992]
```

```python
og_since1992.head()
```

```python
summer.Year.between(1960, 1969).head()
```

```python
og_60s = summer.loc[summer.Year.between(1960, 1969, inclusive=True)]
```

```python
og_60s
```

```python
my_favorite_games = [1972, 1996]
```

```python
summer.Year.isin(my_favorite_games).head()
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_favorite_games)]
```

```python
og_72_96
```

```python
og_not_72_76 = summer.loc[~summer.Year.isin(my_favorite_games)]
```

```python
og_not_72_76
```

```python
og_not_72_76.Year.unique()
```
