```python
import pandas as pd
```

```python
titanic = pd.read_csv("data/titanic.csv")
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot(subplots=True, figsize=(15, 12), sharex=False, sharey=False)
```

```python
titanic.age.plot(figsize=(12, 8), fontsize=13, c='orange', linestyle = '-', 
                 xlim=(0,900), ylim=(0,80), xticks=[x for x in range(0,901, 50)], 
                 yticks=[y for y in range(0, 81, 5)], rot=45)
plt.title("Titanic - Ages", fontsize=15)
plt.legend(loc = "best",fontsize=15)
plt.xlabel("Passenger No", fontsize=13)
plt.ylabel("Age", fontsize=13)
```

```python
plt.style.use("seaborn")
```

```python
titanic.age.value_counts(bins=10)
```

```python
titanic.age.plot(kind="hist", figsize=(12,8), fontsize=15, bins=80)
```

```python
titanic.age.hist(figsize=(12,8), bins=80, xlabelsize=15, ylabelsize=15, cumulative=True)
```

```python
plt.figure(figsize=(12,8))
plt.hist(titanic.age.dropna(), bins=80, density=True, cumulative=True)
```
